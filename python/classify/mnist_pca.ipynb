```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
from scipy.sparse import coo_matrix
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
import cPickle as pickle
```

```python
with open ('/home/jeff/data/mnist/mnist.pkl', 'rb') as f:
    mnist = pickle.load (f)
```

__Compute principal components:__ Using the SVD, we'll compute the principal components of the training data---the vectors that 'best' explain the data, directions with highest variance.

```python
Ndata = 1000
data, rows, cols = [], [], []
for r,d in enumerate (mnist[:Ndata]):
    for c,x in enumerate (d[0]):
        if not x: continue
        
        data.append (x)
        rows.append (r)
        cols.append (c)
shape = (Ndata, len (mnist[0][0]))
```

```python
# construct [nsamples, nfeatures] data matrix
Xsparse = coo_matrix ((data, (rows, cols)), shape=shape, dtype=float)
Xdense = Xsparse.todense ()
```

```python
scaler = StandardScaler (with_mean=True, with_std=True)
Xcenter = scaler.fit_transform (Xdense)
```

```python
U,S,V = np.linalg.svd (Xcenter.T)
```

```python
fig = plt.figure (figsize=(10,6))
for i in range (15):
    u = U[:,i].reshape ((28,28))
    ax = fig.add_subplot (3,5,i+1)
    ax.imshow (u, interpolation='nearest', cmap=plt.cm.binary)
    ax.set_title (('%0.3f' % S[i]))
    ax.axis ('off')
```

```python
plt.plot (S, 'b.', lw=2)
plt.title ('Singular values')
```

__PCA representation:__ Let's look at how the data is distributed by looking at its first two principal components---is each class (digit) easily seperable?

```python
u0, u1 = U[:,0], U[:,1]
```

```python
X1, X2 = np.dot (Xcenter, u0), np.dot (Xcenter, u1)
```

```python
y = [int (d[1]) for d in mnist[:Ndata]]
```

```python
fig = plt.figure (figsize=(9,7))
ax = fig.add_subplot (111)

sc = ax.scatter (X1, X2, c=y, linewidths=0)
fig.colorbar (sc)

ax.axis ('equal')
ax.grid ()
```

Clearly not seperable... Can we do better? Let's try learning a metric that will map data with the same label to be close, but data with different labels to be farther apart.

###Metric learning

We want data that is easily seperable---makes it easier to classify new data.

The (Mahalanobis) distance between two feature vectors, $x_1$ and $x_2$ is

$$
\begin{align}
d_M = (x_1 - x_2)^\top M (x_1 - x_2) .
\end{align}
$$

Here, we'd like to learn a low-rank (positive semi-definite) matrix $M$ that keeps the distance between feature vectors with the same class label close, but provides a larger distance for feature vectors with different class labels. We're shooting for a low-rank matrix so that we'll simultaneously apply dimensionality reduction to the data. In our case, we're looking for a rank 2 matrix.
